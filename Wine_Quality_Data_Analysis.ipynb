```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path1 = '/content/drive/MyDrive/2. Projects/Colab Notebooks/Portfolio/Wine Quality/WineQT.csv'
data_path2 = '/content/drive/MyDrive/2. Projects/Colab Notebooks/Portfolio/Wine Quality/wine-quality-white-and-red.csv'
wine_data = '/content/drive/MyDrive/2. Projects/Colab Notebooks/Portfolio/Wine Quality/Wine.csv'
```

### **Dataset 1**

```python
data1 = pd.read_csv(data_path1)
print('Shape: ', data1.shape, '\n')
print('Columns: ', data1.columns.tolist(), '\n')
data1.head()
```

```python
# Adding new columns
data1['Colour'] = 'Red'
data1['Country'] = 'Portugal'
data1['Name'] = 'Vinho Verde'
```

```python
column_mapping = {
    'fixed acidity': 'Fixed Acidity',
    'volatile acidity': 'Volatile Acidity',
    'citric acid': 'Citric Acid',
    'residual sugar': 'Residual Sugar',
    'chlorides': 'Chlorides',
    'free sulfur dioxide': 'Free Sulfur Dioxide',
    'total sulfur dioxide': 'Total Sulfur Dioxide',
    'density': 'Density',
    'pH': 'pH',
    'sulphates': 'Sulphates',
    'alcohol': 'Alcohol',
    'quality': 'Quality',
    'Id': 'ID',
    'Colour': 'Colour',
    'Country': 'Country',
    'Name': 'Name'
}

# Rename columns using the mapping
data1 = data1.rename(columns=column_mapping)
```

```python
# Define the desired order
desired_order = ['Name', 'Colour', 'Country', 'Fixed Acidity', 'Volatile Acidity', 'Citric Acid', 'Residual Sugar', 'Chlorides', 'Free Sulfur Dioxide', 'Total Sulfur Dioxide', 'Density', 'pH', 'Sulphates', 'Alcohol', 'Quality']

# Reorder columns
data1 = data1[desired_order]
```

```python
print('Shape: ', data1.shape, '\n')
print('Columns: ', data1.columns.tolist(), '\n')
data1.head()
```

### **Dataset 2**

```python
data2 = pd.read_csv(data_path2)
print('Shape: ', data2.shape, '\n')
print('Columns: ', data2.columns.tolist(), '\n')
data2.head()
```

```python
# Adding new columns
data2['Country'] = 'Portugal'
data2['Name'] = 'Vinho Verde'
```

```python
col_old = ['type', 'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
                 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol',
                 'quality', 'Country', 'Name']

col_new = ['Colour', 'Fixed Acidity', 'Volatile Acidity', 'Citric Acid', 'Residual Sugar', 'Chlorides',
                    'Free Sulfur Dioxide', 'Total Sulfur Dioxide', 'Density', 'pH', 'Sulphates', 'Alcohol',
                    'Quality', 'Country', 'Name']

data2 = data2[col_old].rename(columns=dict(zip(col_old, col_new)))
```

```python
# Define the desired order
desired_order = ['Name', 'Colour', 'Country', 'Fixed Acidity', 'Volatile Acidity', 'Citric Acid', 'Residual Sugar', 'Chlorides', 'Free Sulfur Dioxide', 'Total Sulfur Dioxide', 'Density', 'pH', 'Sulphates', 'Alcohol', 'Quality']

# Reorder columns
data1 = data1[desired_order]
```

```python
print('Shape: ', data2.shape, '\n')
print('Columns: ', data2.columns.tolist(), '\n')
data2.head()
```

```python
data = pd.concat([data1, data2], ignore_index=True)
print('Shape: ', data.shape, '\n')
print('Columns: ', data.columns.tolist(), '\n')
data.head()
```

```python
# Assuming 'combined_data' is the combined DataFrame from the previous example
data['Colour'] = data['Colour'].replace({'red': 'Red', 'white': 'White'}, regex=True, inplace=False)
```

```python
for column_name, values in data.items():
    unique_values = set(values)
    num_unique_values = len(unique_values)

    print(f"Column: {column_name}")
    print(f"Unique Values: {unique_values}")
    print(f"Number of Unique Values: {num_unique_values}\n")
```

## **Combine 2 datasets**

```python
data.to_csv(wine_data, index=False)

# Confirm the file is saved
print("Wine.csv saved successfully.")
```
